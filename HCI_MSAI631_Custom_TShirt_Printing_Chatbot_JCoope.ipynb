## 6. Traditional Chatbot Decision Logic

The chatbot uses Python decision rules to recommend the most appropriate printing method.

```
# ===========================================================
# CHATBOT DECISION FUNCTION
# ===========================================================

def recommend_printing(event):

    event = event.lower().strip()

    # Check for blank input
    if event == "":
        print("\nSorry.")
        print("Please enter an event type.")
        return

    # Check if the event exists in the knowledge base
    if event in printing_options:

        recommendation = printing_options[event]

        print("\n====================================")
        print(" CUSTOM T-SHIRT RECOMMENDATION")
        print("====================================")

        print(f"\nEvent: {event.title()}")
        print(f"Printing Method : {recommendation['printing_method']}")
        print(f"Shirt Material  : {recommendation['shirt_material']}")
        print(f"Best For        : {recommendation['best_for']}")
        print(f"Production Time : {recommendation['production']}")

    else:

        print("\nSorry.")
        print("I don't have a recommendation for that event yet.")
```

# 10. AI Recommendation Assistant

This section integrates the traditional rule-based chatbot with the Hugging Face cloud-based AI service. After the chatbot provides a recommendation, the AI generates additional explanations and creative suggestions, including color ideas, font recommendations, design concepts, and shirt care instructions.

```
# ===========================================================
# AI PROMPT GENERATOR
# ===========================================================

def create_ai_prompt(event):

    # Get the chatbot recommendation
    recommendation = printing_options[event]

    prompt = f"""
You are a professional custom T-shirt printing consultant.

A customer is ordering custom T-shirts for a {event} event.

The traditional rule-based chatbot has already recommended:

Printing Method: {recommendation['printing_method']}
Shirt Material: {recommendation['shirt_material']}
Best For: {recommendation['best_for']}
Production Time: {recommendation['production']}

Do NOT recommend a different printing method.

Instead, explain why this recommendation is appropriate.

Also provide:

1. Two shirt color recommendations.
2. One font recommendation.
3. One creative design idea.
4. Two shirt care tips.

Keep the response friendly and under 200 words.
"""

    return prompt
```

```
print(create_ai_prompt("birthday"))
```

```
# ===========================================================
# GET AI RECOMMENDATION FROM HUGGING FACE
# ===========================================================

def get_ai_recommendation(event):

    prompt = create_ai_prompt(event)

    response = client.chat.completions.create(
        model="Qwen/Qwen2.5-7B-Instruct",
        messages=[
            {
                "role": "user",
                "content": prompt
            }
        ],
        max_tokens=250
    )

    return response.choices[0].message.content
```

```
print(get_ai_recommendation("birthday"))
```

```
# ===========================================================
# DISPLAY FORMATTED AI RECOMMENDATION
# ===========================================================

def display_ai_recommendation(event):

    print("\n")
    print("=" * 60)
    print("               AI DESIGN ASSISTANT")
    print("=" * 60)

    response = get_ai_recommendation(event)

    print(response)

    print("=" * 60)
    print("End of AI Recommendation")
    print("=" * 60)
```

```
display_ai_recommendation("birthday")
```

# 11. Gradio User Interface

This section creates a simple web-based interface using Gradio. The interface allows users to interact with the chatbot by selecting an event type from a drop-down menu. The traditional chatbot generates the recommendation, while the Hugging Face AI service provides additional explanations and design suggestions.

```
# ===========================================================
# GRADIO CHATBOT FUNCTION
# ===========================================================

def chatbot_interface(event):

    recommendation = printing_options.get(event.lower())

    if recommendation is None:
        return "Sorry. This event is not currently supported."

    chatbot_result = f"""
CUSTOM T-SHIRT RECOMMENDATION

Event: {event.title()}

Printing Method: {recommendation['printing_method']}
Shirt Material: {recommendation['shirt_material']}
Best For: {recommendation['best_for']}
Production Time: {recommendation['production']}
"""

    ai_result = get_ai_recommendation(event)

    return chatbot_result + "\n\n" + "="*60 + "\nAI DESIGN ASSISTANT\n" + "="*60 + "\n\n" + ai_result
```

```
# ===========================================================
# BUILD GRADIO INTERFACE
# ===========================================================

demo = gr.Interface(
    fn=chatbot_interface,

    inputs=gr.Dropdown(
        choices=[
            "birthday",
            "church",
            "business",
            "family",
            "sports",
            "school",
            "fundraiser",
            "personal"
        ],
        label="Select Event Type"
    ),

    outputs=gr.Textbox(
        lines=25,
        label="Chatbot Recommendation"
    ),

    title="Custom T-Shirt Printing Chatbot",

    description="""
Traditional Rule-Based Chatbot Integrated with a Cloud-Based AI Service

Select an event type to receive a printing recommendation and AI-powered design suggestions.
"""
)
```

```
# ===========================================================
# LAUNCH CHATBOT
# ===========================================================

demo.launch()
```
